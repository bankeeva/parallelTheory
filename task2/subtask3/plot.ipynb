```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 8, 16, 20, 26, 32, 40, 57, 64, 73, 80])
y1 = np.array([1.96923, 3.55287, 6.50169, 13.2978, 13.2977, 14.6104, 15.4233, 16.2117, 12.0721, 15.3094, 15.2303, 12.7115]) / x
y2 = np.array([1.87227, 3.4961, 6.747, 12.39, 15.2446, 16.9027, 19.6047, 22.3505, 17.0533, 16.5478, 28.158, 20.9674]) / x

print("График эффективности") # ускорение / потоки
plt.plot(x, y1, 'red', label='v1')
plt.plot(x, y2, 'green', label='v2')
plt.xlabel('p')
plt.ylabel('Ep')
plt.legend()
plt.grid(True)
plt.show()
```

```python
labels = ["static,0", "static,1", "static,10", "static,100", "static,250",
    "dynamic,0", "dynamic,1", "dynamic,10", "dynamic,100", "dynamic,250",
    "guided,0", "guided,1", "guided,10", "guided,100", "guided,250"
]

v1 = [
    0.323366, 0.438355, 0.426179, 0.426218, 0.285151,
    0.871388, 0.864255, 0.57885, 0.528163, 0.551257,
    0.524243, 0.52947, 0.488416, 0.519789, 0.510102
]

v2 = [
    0.80351, 0.886888, 0.771998, 0.893791, 0.743622,
    1.19299, 1.27308, 0.926349, 0.998209, 0.926635,
    0.914033, 0.906898, 0.889763, 0.943619, 0.831679
]

x = np.arange(len(labels))
width = 0.38

plt.figure(figsize=(16, 7))

bars1 = plt.bar(x - width/2, v1, width, label='V1')
bars2 = plt.bar(x + width/2, v2, width, label='V2')

plt.title("Исследование параметров schedule\nN = 2000, num_threads = 8")
plt.xlabel("Параметры schedule")
plt.ylabel("Время выполнения, с")
plt.xticks(x, labels, rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

for bar in bars1:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h + 0.01, f"{h:.3f}",
             ha='center', va='bottom', fontsize=7)

for bar in bars2:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h + 0.01, f"{h:.3f}",
             ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()
```
