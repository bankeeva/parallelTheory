```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 7, 8, 16, 20, 40])
speed2 = np.array([1.99468, 4.09493, 7.68032, 8.77037, 10.03, 10.5628, 12.4404])
speed4 = np.array([2.38431, 5.09272, 5.9162, 5.87336, 6.96223, 7.40207, 10.3381])

x_line = np.array([0, 40])

plt.plot(x, speed2, 'red', label='M = 20_000')
plt.plot(x, speed4, 'green', label='M = 40_000')
plt.plot(x_line, x_line, 'b--', label='y = x')
plt.xlabel('p')
plt.ylabel('Sp')
plt.legend()
plt.grid(True)
plt.show()
```

```python
effectiv2 = speed2 / x
effectiv4 = speed4 / x

plt.plot(x, effectiv2, 'red', label='20_000')
plt.plot(x, effectiv4, 'green', label='40_000')
plt.xlabel('p')
plt.ylabel('Ep')
plt.legend()
plt.grid(True)
plt.show()
```

```python
times2 = np.array([0.201686, 0.0982434, 0.0523807, 0.0458703, 0.0401097, 0.0380864, 0.0323382])
times4 = np.array([0.858077, 0.401734, 0.345816, 0.348339, 0.29386, 0.276398, 0.197901])

def optimal_threads(x, times, effectiv):
    optimal_th = []
    for i in range(len(x) - 1):
        if effectiv[i] >= 0.8 and times[i] <= sorted(times)[-3]:
            optimal_th.append(x[i])
    
    return optimal_th

print(f"20_000: {max(optimal_threads(x, times2, effectiv2))}")
print(f"40_000: {max(optimal_threads(x, times4, effectiv4))}")
```
