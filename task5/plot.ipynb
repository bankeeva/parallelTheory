```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 4, 6, 8, 16])
time = np.array([40.06175441600499, 28.94040799999493, 22.5731957500102, 23.773281749978196, 25.84879550000187, 45.76709333300823])

speed = time / 40.06175441600499

x_line = np.array([0, 16])

plt.plot(x, speed, 'red')
plt.plot(x_line, x_line, 'b--', label='y = x')
plt.xlabel('p')
plt.ylabel('Sp')
plt.legend()
plt.grid(True)
plt.show()
```
