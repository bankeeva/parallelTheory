```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sizes = ['128x128', '256x256', '512x512', '1024x1024']
multi = [51.6721, 106.987, 902.352, 2113.02]
cpu = [9.6803, 43.3251, 145.613, 622.314]

x = range(len(sizes))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar([p - width/2 for p in x], multi, width, label='Multicore80', color='skyblue')
plt.bar([p + width/2 for p in x], cpu,  width, label='CPU', color='red', alpha=0.8)

plt.title('Сравнение CPU и Multicore')
plt.xlabel('Размер сетки')
plt.ylabel('Время (секунды)')
plt.xticks(x, sizes)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
stages = ['1', '2', '3', '4', '5', '6']
times = [23.0737, 10.8397, 10.009, 7.70377, 4.49717, 4.88431]

plt.figure(figsize=(8, 5))
plt.bar(stages, times, color='skyblue')

plt.title('Этапы оптимизации (512×512)')
plt.xlabel('Этап №')
plt.ylabel('Время выполнения (секунды)')

for i, v in enumerate(times):
    plt.text(i, v + 0.5, f'{v:.2f}', ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
sizes = ['128x128', '256x256', '512x512', '1024x1024']
gpu = [0.199005, 0.738572, 4.88431, 25.5937]
multi = [51.6721, 106.987, 902.352, 2113.02]
cpu = [9.6803, 43.3251, 145.613, 622.314]

x = range(len(sizes))
width = 0.25

plt.figure(figsize=(8, 5))

plt.bar([p - width for p in x], cpu,   width=width, label='CPU', color='red')
plt.bar([p for p in x], multi, width=width, label='Multicore80', color='blue')
plt.bar([p + width for p in x], gpu,   width=width, label='GPU', color='green')

plt.title('CPU vs Multicore vs GPU')
plt.xlabel('Размер сетки')
plt.ylabel('Время (секунды)')
plt.xticks(x, sizes)
plt.legend()
plt.grid(axis='y', linestyle='--')

plt.show()
```
